## 前處理
- 載入股票
- 轉換數值

```python
def trans_int(input):
    try:
        return int(input)
    except:
        print(input)
        return None
    
def trans_float(input):
    try:
        return float(input)
    except:
        print(input)
        return None
```

```python
def preprocess(df):
    df["收盤價"] = df["收盤價"].map(trans_float)
    df["最高價"] = df["最高價"].map(trans_float)
    df["最低價"] = df["最低價"].map(trans_float)
    return df
df = preprocess(df)
```
